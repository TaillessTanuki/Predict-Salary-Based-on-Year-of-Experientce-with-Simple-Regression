# Simple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
# pd = pandas, read_csv = function that can read csv file that we have in the folder
dataset = pd.read_csv("Salary_Data.csv")

# creating matrix fitures/predictor (independent variables) and dependent variables
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
# in the video at 6:23
x_train, x_test, y_train, y_test =  train_test_split(x,y,test_size=1/3,random_state=0)
```

## Training the Simple Linear Regression model on the Training set

```python
# 1. import the right class from library sklearn, linear modell, linear regression
# and make an object call regressor (object that will regressed the data)
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

# 2. Connect the regressor to the training sets with fit method (.fit(training sets))
regressor.fit(x_train,y_train)

# regressor is an object that contain trained x and y, so it can used to predict the test set
```

## Predicting the Test set results

```python
# We want to predict the salary of the employee based on their experience, using
# predict method (.predict())
# so y prediction = regressor.predict(x test set)
y_pred = regressor.predict(x_test) 

# note : y_test contains real salaries, y_predict contains predicted salaries

```

## Visualising the Training set results

```python
# compare the real salaries to the predict salaries (training set and predict)
# calling plt that stands for mathplotlib
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')

# makin chart 
plt.title("Salary vs Experience (Training Set)")
plt.xlabel('Years of Experiences')
plt.ylabel('Salary')
plt.show()

# Note : this is for Just Training set (not Test Set)
```

## Visualising the Test set results

```python
# compare the real salaries to the predict salaries (test set and predict)
# calling plt that stands for mathplotlib
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')

# makin chart 
plt.title("Salary vs Experience (Test Set)")
plt.xlabel('Years of Experiences')
plt.ylabel('Salary')
plt.show()

# Note : this is for Just Test set (not Training Set)
# Note : but in the regression plt.plot(x_test, regressor.predict(x_test), color = 'blue'),
# is the same with plt.plot(x_train, regressor.predict(x_train), color = 'blue')
# because the regression line came from data in independet varibles from the table 
# in this case x_train
```

## Making a single prediction (for example the salary of an employee with 12 years of experience)


```python
# We want to predict salary of an employee that has 12 years of experience
# to do that we using object regressor and predict method
regressor.predict([[12]])

# Notice that the we using double bracket in the predict input, it is because predict expects a 2d array as format of its input.
# simply put : 
# 12→scalar
# [12]→1D array
# [[12]]→2D array
```

So the employee with 12 years of experience get 138.967,5 Salary


## Getting the final linear regression equation with the values of the coefficients

```python
# Getting the final linear regression function equation (yhat=ax+b)
# 1. Getting the coeficient of the x (a)
print(regressor.coef_)

# 2. Getting the constant b
print(regressor.intercept_)
```
